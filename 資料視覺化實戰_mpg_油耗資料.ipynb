# Geoms + Stats ÂØ¶‰æã
bar charts, histograms: Ë®àÁÆóÊØè‰∏ÄÁµÑ bin Ë£°Èù¢ÁöÑÊï∏ÁõÆ.

```r
library(dplyr)
d <- mpg %>% 
  group_by(class) %>% 
  summarise(n = n())
d
ggplot(data = d) +
  geom_bar(mapping = aes(x = reorder(class, -n), y = n),
           stat = "identity")
```

```r
ggplot(data = mpg, aes(x = class)) +
  geom_bar() +
  geom_text(stat = "count",
            aes(label = ..count.., y =..count..),
            vjust = "bottom")
```

# ËÆìË¶ñË¶∫ÂåñÂ¢ûÂä†Ë≥áË®äÁöÑÊ∏ÖÊô∞Â∫¶: facet_wrap()

Áï´Âá∫ÂêÑËªäÂª†(manufacturer)Âú®‰∏çÂêåËªäÂûã(class)ÁöÑÊï∏Èáè bar chart

x: class
y: count
fill: Â§öÈ§òÁöÑ mappingÔºå‰ΩÜÁÇ∫‰∫ÜÂº∑Ë™ø‰∏çÂêåËªäÂûãÔºåÈÇÑÊòØÂä†ÂÖ•
facet: manufacturer

```r

ggplot(data = mpg) +
  geom_bar(mapping = aes(x = class, fill = class)) +
  facet_wrap( ~ manufacturer, ncol = 4) +
  scale_fill_viridis_d() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

```r
mpg %>% 
  group_by(manufacturer, class) %>%
  tally() %>% 
  ggplot(aes(x = class, y = manufacturer, fill = n)) +
  geom_tile() +
  geom_text(aes(label = n)) +
  scale_fill_distiller(direction = 1) +
  scale_x_discrete(position = "top") +
  ggtitle("Number of car class by manufacturers")
```
